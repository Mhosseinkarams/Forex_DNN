# DNN Multi InOut - Sequence Model Testing
This notebook tests the LSTM-based classifiers for sequence-based price prediction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from pathlib import Path

# Add parent directory to path
sys.path.append(str(Path.cwd()))

from classifier_binary import BinaryLSTMClassifier
from classifier_multi import MultiLSTMClassifier
from classifier_pivot import PivotLSTMClassifier
```

```python
# Test BinaryLSTMClassifier
bin_lstm = BinaryLSTMClassifier(num_input_candles=24) # Shorter for test
X, y = bin_lstm.load_and_preprocess()
if X is not None:
    history, eval_metrics = bin_lstm.train(X, y, epochs=5, batch_size=128)
    print(f"Binary LSTM Test Accuracy: {eval_metrics[1]:.4f}")
```

```python
def plot_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='val')
    plt.title(f'{title} Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='val')
    plt.title(f'{title} Loss')
    plt.legend()
    plt.show()

if 'history' in locals():
    plot_history(history, "Binary LSTM")
```

```python
# Test MultiLSTMClassifier
multi_lstm = MultiLSTMClassifier(num_input_candles=24)
X_m, y_m = multi_lstm.load_and_preprocess()
if X_m is not None:
    history_m, eval_m = multi_lstm.train(X_m, y_m, epochs=5, batch_size=128)
    plot_history(history_m, "Multi-class LSTM")
```

```python
# Test PivotLSTMClassifier
pivot_lstm = PivotLSTMClassifier(num_input_candles=24)
X_p, y_p = pivot_lstm.load_and_preprocess()
if X_p is not None:
    history_p, eval_p = pivot_lstm.train(X_p, y_p, epochs=5, batch_size=128)
    plot_history(history_p, "Pivot LSTM")
```
