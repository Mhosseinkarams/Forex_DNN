# DNN Single InOut - Model Testing
This notebook tests the binary and multi-class DNN classifiers for price prediction.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
from pathlib import Path

# Add parent directory to path
sys.path.append(str(Path.cwd()))

from classifier_binary import BinaryDNNClassifier
from classifier_multi import MultiDNNClassifier
```

```python
# Test BinaryDNNClassifier
bin_classifier = BinaryDNNClassifier()
X, y = bin_classifier.load_data()
if X is not None:
    history, eval_metrics = bin_classifier.train(X, y, epochs=10) # Short train for test
    print(f"Test Loss: {eval_metrics[0]:.4f}, Test Accuracy: {eval_metrics[1]:.4f}")
```

```python
# Plot Training History
def plot_history(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='val')
    plt.title('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='val')
    plt.title('Loss')
    plt.legend()
    plt.show()

if 'history' in locals():
    plot_history(history)
```

```python
# Test MultiDNNClassifier
multi_classifier = MultiDNNClassifier()
X_m, y_m = multi_classifier.load_data()
if X_m is not None:
    history_m, eval_metrics_m = multi_classifier.train(X_m, y_m, epochs=10)
    print(f"Multi-class Test Loss: {eval_metrics_m[0]:.4f}, Test Accuracy: {eval_metrics_m[1]:.4f}")
    plot_history(history_m)
```
